```python
from sklearn.cross_validation import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn import tree 
#from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
import pandas as pd 
#import numpy as np 

```

```python
#import pandas as pd 
datasets = pd.read_csv('playTennis.csv')
datasets
```

```python
ab=preprocessing.LabelEncoder()
datasets=datasets.apply(ab.fit_transform)
datasets
```

```python
#split dataset in to featurs and target variables
feature_cols=['Outlook','Temperature', 'Humidity','Wind']
X = datasets[feature_cols] # Features
Y = datasets.Play # Target variable
X
```

```python
Y
```

```python
##split dataset in to train set and test set
X_train, X_test , Y_train,Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
# cerat descision tree classifeir object--clf
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train.astype(int),Y_train.astype(int)) 
Y_pred_en = clf_entropy.predict(X_test)

Y_pred_en
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy is :{0}".format(accuracy_score(Y_test.astype(int),Y_pred_en) * 100))
```
